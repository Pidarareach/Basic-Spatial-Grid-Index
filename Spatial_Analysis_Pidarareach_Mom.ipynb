The basic spatial indexing technique--spatial grid index.

1. Generate 10K random points in the two-dimensional space. Set the range of x and y between 0
and 100.
2. Construct the grid and map each of the random points into the grid.

```python
# Grid
# Size of grid (n,m)
n, m = 3, 4

# Size of each grid (A,B)
A, B = 3, 4

# the number of grids
num_grids = (n // A) * (m // B)
print(num_grids)
```

```python
# Dictionary to store points in grid
grid_list = {}

import random

# Generate 10k random points
for i in range(10000):
     x = random.randint(0, 100)
     y = random.randint(0, 100)

     # calculates the x and y index of the grid based on the x and y coord
     grid_x = x // A
     grid_y = y // B

     # create a tuple representing the grid index
     grid = (grid_x, grid_y)

    # checks if grid index is already created, if not adds a new list
     if grid not in grid_list:
      grid_list[grid] = []

    # append (x,y) to the dictionary
     grid_list[grid].append((x, y))

# Display
for grid, points in grid_list.items():
  print(f'Grid {grid}: {points}')
```

```python
# Visualization
import matplotlib.pyplot as plt

# plot points in each grid
for grid, points in grid_list.items():
  grid_x, grid_y = grid   # unpack
  x_coords, y_coords = zip(*points)   # gets (x,y) coords from the points
  plt.scatter(x_coords, y_coords)

plt.grid()
plt.show()
```
